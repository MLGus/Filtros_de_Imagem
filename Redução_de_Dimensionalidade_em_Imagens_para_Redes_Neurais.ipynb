```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload do arquivo
uploaded = files.upload()

# Pegando o nome do arquivo enviado
caminho_imagem = list(uploaded.keys())[0]

# Função para processar a imagem
def processar_imagem(caminho_imagem, limiar=128):
    # Carregar a imagem colorida
    imagem_colorida = cv2.imread(caminho_imagem)
    imagem_colorida = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

    # Converter para níveis de cinza
    imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_RGB2GRAY)

    # Binarizar a imagem
    _, imagem_binarizada = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)

    # Exibir as imagens
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].imshow(imagem_colorida)
    axs[0].set_title("Imagem Colorida")
    axs[0].axis("off")

    axs[1].imshow(imagem_cinza, cmap="gray")
    axs[1].set_title("Imagem em Níveis de Cinza")
    axs[1].axis("off")

    axs[2].imshow(imagem_binarizada, cmap="gray")
    axs[2].set_title("Imagem Binarizada")
    axs[2].axis("off")

    plt.show()

# Chamar a função com a imagem carregada
processar_imagem(caminho_imagem)
```
